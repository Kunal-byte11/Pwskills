# Statistics Advance Part 1

## 1. What is a Random Variable?
A function that assigns numerical values to outcomes of a random experiment.

## 2. Types of Random Variables
- **Discrete:** Finite/countable values (e.g., number of heads)
- **Continuous:** Any value in a range (e.g., height, weight)

## 3. Discrete vs Continuous Distributions
| Aspect | Discrete | Continuous |
|--------|----------|-----------|
| Values | Countable | Any in interval |
| Probability | P(X=x) | P(a≤X≤b) via integral |
| Graph | Bar/step | Smooth curve |

## 4. Probability Distribution Function (PDF)
Shows probability of each outcome (discrete) or probability density (continuous). For discrete: sum=1. For continuous: integral under curve=1.

## 5. CDF vs PDF
- **PDF (f(x)):** Probability at specific value
- **CDF (F(x)):** Probability up to that value → F(x) = P(X ≤ x)
- **Relationship:** CDF is cumulative sum/integral of PDF

## 6. Discrete Uniform Distribution
All outcomes equally likely. P(X=x) = 1/n where n = number of outcomes.
- **Mean:** E(X) = (a+b)/2
- **Variance:** (n²-1)/12

## 7. Bernoulli Distribution
Single trial with two outcomes (success/failure).
- **P(X=x) = p^x(1-p)^(1-x)** for x ∈ {0,1}
- **Mean:** p | **Variance:** p(1-p)

## 8. Binomial Distribution
Number of successes in n independent trials.
- **P(X=x) = C(n,x)p^x(1-p)^(n-x)**
- **Mean:** np | **Variance:** np(1-p)
- **Uses:** Quality control, clinical trials, surveys

## 9. Poisson Distribution
Number of events in fixed interval at constant rate λ.
- **P(X=x) = (e^(-λ)λ^x)/x!**
- **Mean = Variance = λ**
- **Uses:** Call arrivals, defects, web traffic

## 10. Continuous Uniform Distribution
Equal probability across interval [a,b].
- **f(x) = 1/(b-a)**
- **Mean:** (a+b)/2 | **Variance:** (b-a)²/12

## 11. Normal Distribution Characteristics
- Bell-shaped, symmetric around mean μ
- Completely defined by μ and σ
- **68-95-99.7 rule:** 68% within 1σ, 95% within 2σ, 99.7% within 3σ

## 12. Standard Normal Distribution
Normal distribution with μ=0, σ=1. Denoted Z ~ N(0,1).
**Why important:** Standardizes any normal distribution, enables probability calculations without tables.

## 13. Central Limit Theorem (CLT)
Distribution of sample means approaches normal distribution as n increases, regardless of original population distribution.
- **Why critical:** Enables statistical inference without knowing population distribution
- **Condition:** n ≥ 30 (general rule)

## 14. CLT & Normal Distribution Relationship
CLT proves sample means follow normal distribution even if population isn't normal. This connection allows using normal distribution theory for inference.

## 15. Z-Statistics in Hypothesis Testing
Test whether population mean differs from hypothesized value.
- **Z = (X̄ - μ₀)/(σ/√n)**
- **Decision:** If |Z| > critical value → reject null hypothesis

## 16. Z-Score: Calculation & Meaning
**Z = (X - μ)/σ** → Measures how many standard deviations from mean.
- Z=1 means 1 SD above mean
- Z=-2 means 2 SD below mean
- Allows comparison across different distributions

## 17. Point vs Interval Estimates
- **Point Estimate:** Single value (e.g., sample mean = 75)
- **Interval Estimate:** Range (e.g., 70-80 with 95% confidence)
- Interval is more realistic; accounts for sampling variability

## 18. Significance of Confidence Intervals
Shows range where population parameter likely lies with specified confidence.
- **Example:** "95% confident mean is between 70-80"
- Quantifies uncertainty in estimates
- Basis for statistical inference decisions

## 19. Z-Score & Confidence Interval Relationship
Z-score provides **critical value** for confidence interval formula:
**CI = Estimate ± Z × Standard Error**
- Z from standard normal table based on confidence level
- Higher confidence → larger Z → wider interval

## 20. Using Z-Scores to Compare Distributions
Standardizes values to common scale (mean=0, σ=1).
- Score 85 on test (μ=75, σ=10): Z = 1.0
- Score 88 on test (μ=75, σ=5): Z = 2.6
- Second score is relatively better (2.6 > 1.0)

## 21. CLT Assumptions
1. **Independence:** Observations independent
2. **Identical Distribution:** All from same population
3. **Finite Variance:** Population variance exists
4. **Sufficient Sample Size:** n ≥ 30 generally

## 22. Expected Value
Long-run average of random variable.
**E(X) = Σ[x × P(X=x)]** (discrete) or **∫ x·f(x)dx** (continuous)
- Fair lottery price = expected winnings
- Average outcome over many trials

## 23. Probability Distribution & Expected Outcome
Distribution describes all possible outcomes & probabilities. Expected value is the weighted average from that distribution.
- Distribution = complete information
- Expected value = summary of distribution
- Different distributions can have same expected value but different spreads

## 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)>

```python
import numpy as np
import matplotlib.pyplot as plt

# Define range (example: 1 to 6 like a dice)
a = 1
b = 6

# Generate values
x = np.arange(a, b + 1)

# PMF: equal probability for all values
pmf = np.ones_like(x) / len(x)

# Plot PMF
plt.figure()
plt.stem(x, pmf)
plt.xlabel("Values")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution PMF")
plt.show()
```

## 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution

    Parameters:
    x : int (0 or 1)
    p : probability of success (0 <= p <= 1)

    Returns:
    Probability value
    """

    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0  # Invalid input


# Example usage
p = 0.7

print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

## 6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution>

```python
import numpy as np
import matplotlib.pyplot as plt

# Define range (example: 1 to 6)
a = 1
b = 6

# Generate values
x = np.arange(a, b + 1)

# PMF (equal probability)
pmf = np.ones_like(x) / len(x)

# CDF (cumulative sum of PMF)
cdf = np.cumsum(pmf)

# Plot CDF
plt.figure()
plt.step(x, cdf, where='post')
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Discrete Uniform Distribution")
plt.ylim(0, 1.1)
plt.show()
```

## 9. Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset
    """
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std
    return z_scores

# Example dataset
data = np.array([10, 12, 15, 18, 20, 22, 25])

# Calculate Z-scores
z = calculate_z_scores(data)

print("Z-scores:", z)

# Plot Z-scores
plt.figure()
plt.plot(z, marker='o')
plt.axhline(0)  # mean line
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.title("Z-score Plot")
plt.show()
```

## 10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate non-normal data (Uniform distribution)
population = np.random.uniform(0, 10, 10000)

# Step 2: Function to compute sample means
def sample_means(population, sample_size, num_samples):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, sample_size)
        means.append(np.mean(sample))
    return means

# Step 3: Generate sample means
sample_size = 30
num_samples = 1000

means = sample_means(population, sample_size, num_samples)

# Step 4: Plot histogram of sample means
plt.figure()
plt.hist(means, bins=30, density=True)
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("CLT: Distribution of Sample Means")
plt.show()
```

## 17. Generate random variables and calculate their corresponding probabilities using the binomial distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 10   # number of random samples

# Step 1: Generate random variables (binomial samples)
samples = np.random.binomial(n, p, size)
print("Random Variables:", samples)

# Step 2: Function to calculate Binomial PMF
def binomial_pmf(x, n, p):
    return comb(n, x) * (p**x) * ((1 - p)**(n - x))

# Step 3: Calculate probabilities for each sample
probabilities = [binomial_pmf(x, n, p) for x in samples]

print("Corresponding Probabilities:", probabilities)

# Step 4: Plot PMF for all possible values
x_vals = np.arange(0, n+1)
pmf_vals = [binomial_pmf(x, n, p) for x in x_vals]

plt.figure()
plt.stem(x_vals, pmf_vals)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution PMF (n=10, p=0.5)")
plt.show()
```

## Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from scipy.stats import norm

# Sample dataset
data = np.array([52, 55, 58, 60, 62, 65, 67])

# Given population parameters
mu = 60        # population mean (H0)
sigma = 5      # population standard deviation
n = len(data)

# Step 1: Calculate sample mean
x_bar = np.mean(data)

# Step 2: Calculate Z-statistic
z = (x_bar - mu) / (sigma / sqrt(n))

print("Sample Mean:", x_bar)
print("Z-statistic:", z)

# Step 3: Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z)))

print("P-value:", p_value)

# Step 4: Decision (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypothesis (H0)")
else:
    print("Fail to Reject Null Hypothesis (H0)")

# Step 5: Plot Standard Normal Distribution
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x)

plt.figure()
plt.plot(x, pdf)

# Mark Z value
plt.axvline(z)

plt.title("Z-Test Visualization")
plt.xlabel("Z")
plt.ylabel("Density")
plt.show()
```

## 20. Create a confidence interval for a dataset using Python and interpret the result


```python
import numpy as np
from scipy.stats import norm

# Sample dataset
data = np.array([52, 55, 58, 60, 62, 65, 67])

# Confidence level
confidence = 0.95

# Step 1: Calculate sample statistics
mean = np.mean(data)
std = np.std(data, ddof=1)   # sample standard deviation
n = len(data)

# Step 2: Z-value for confidence level
z = norm.ppf((1 + confidence) / 2)

# Step 3: Margin of error
margin = z * (std / np.sqrt(n))

# Step 4: Confidence interval
lower = mean - margin
upper = mean + margin

print("Sample Mean:", mean)
print("Confidence Interval:", (lower, upper))
```

## 23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter
lam = 4   # average rate (λ)

# Values
x = np.arange(0, 15)

# Calculate CDF
cdf = poisson.cdf(x, lam)

# Print values
for i in range(len(x)):
    print(f"P(X ≤ {x[i]}) = {cdf[i]:.4f}")

# Plot CDF
plt.figure()
plt.step(x, cdf, where='post')
plt.xlabel("Number of Events")
plt.ylabel("Cumulative Probability")
plt.title("Poisson CDF (λ = 4)")
plt.ylim(0, 1.1)
plt.show()
```

27.Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Original dataset
data = np.array([10, 12, 15, 18, 20, 22, 25])

# Step 2: Calculate mean and standard deviation
mean = np.mean(data)
std = np.std(data)

# Step 3: Z-score normalization
z_scores = (data - mean) / std

print("Original Data:", data)
print("Z-score Normalized Data:", z_scores)

# Step 4: Visualization
plt.figure()

# Original data plot
plt.subplot(1, 2, 1)
plt.plot(data, marker='o')
plt.axhline(mean)
plt.title("Original Data")

# Normalized data plot
plt.subplot(1, 2, 2)
plt.plot(z_scores, marker='o')
plt.axhline(0)
plt.title("Z-score Normalized Data")

plt.show()
```

## 28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skew_kurtosis(data):
    """
    Calculate skewness and kurtosis of a dataset
    """
    skewness = skew(data)
    kurt = kurtosis(data)
    return skewness, kurt

# Step 1: Generate normal distribution data
mu = 0
sigma = 1
size = 1000

data = np.random.normal(mu, sigma, size)

# Step 2: Calculate skewness and kurtosis
skewness, kurt = calculate_skew_kurtosis(data)

print("Skewness:", skewness)
print("Kurtosis:", kurt)
```
